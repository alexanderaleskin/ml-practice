### 1. Реализовать ЕМ-алгоритм для восстановления смеси многомерных нормальных распределений. 
Требования по эффективности реализации: среднее время одной итерации не должно превышать одной секунды.

#### Функция, выполняющая ЕМ-алгоритм. 
На вход принимает:

1. матрицу $ X_{(N, d)} $, где $ N $ - число элементов, $ d $ - размерность пространства,
2. список из словарей start_vals, где каждый словарь описывает 1 начальное приближение, 
3. точность вычислений.

Выход: словарь из:

1. списка словарей из параметров $ w $, $\mu $ , $\sum$, количества итераций, значения логарифма правдоподобия и принадлежность элемента к классу.
2. наилучшее приближение

```python
def EM(X, start_vals, tol = 1e-8, max_iter = 40):
    answer = []
    best = {'log_like': -1e30}
    gaussians = start_vals[0]['w'].shape[0]
    
    d = start_vals[0]['means'].shape[1]
    for i in range(len(start_vals)):
        # при 1 гауссине или единичной размерности неправильно оптимизирует и никак не исправить.
        # поэтому приведение к типу float64 надо делать самому!
        if (gaussians == 1):
            w = start_vals[i]['w'].copy()
        else:
            w = np.float64(start_vals[i]['w']).copy()
        if (d == 1 and gaussians == 1):
            means = start_vals[i]['means'].copy()
            stds  = start_vals[i]['stds'].copy()
        else:
            means = np.float64(start_vals[i]['means']).copy()
            stds  = np.float64(start_vals[i]['stds']).copy()
        
        #gaussians = w.shape[0]
        N = X.shape[0]
        dets = np.zeros((1, gaussians))
        exps = np.zeros((N, gaussians))
        I = np.eye(stds.shape[1])
        I2 = I * 1e-4
        work = True
        l_last = -1e30 * N
        l_curr = 0.0
        it = 0
        while work:
            for k in range(gaussians):
                L = np.linalg.cholesky(stds[k])
                dets[0, k] = np.multiply.reduce(L.diagonal()) #** 2
                #print(means[k], L.T)
                exps[:, k] = - 0.5 * np.sum((X - means[k]).dot(sp.linalg.solve_triangular(L.T, I)) ** 2, axis = 1)
                
            maxs = np.max(exps, axis = 1)[np.newaxis].T
            #coefs = np.exp(exps - maxs)* w / dets
            coefs = exps + np.log(w) - np.log(dets) - maxs
            coefs = np.exp(coefs)
            sum_coefs = np.sum(coefs, axis = 1)[np.newaxis].T
            coefs = coefs / (sum_coefs + 1e-5)
            w = np.sum(coefs, axis = 0)
            for k in range(gaussians):
                means[k] = np.sum(coefs[:, k][np.newaxis].T * X, axis = 0) / (w[k] + 1e-5)
                dX = X - means[k]
                stds[k] = (dX.T.dot(coefs[:, k][np.newaxis].T * dX) / (w[k] + 1e-5)) + I2
            w = w / N
            it = it + 1
            l_curr = np.sum(np.log(sum_coefs) + maxs, axis = 0)
            if l_curr < l_last:
                break
            #    raise Exception('Log-likelihood decreased on', it, 'iteration!')
            work = (l_curr - l_last > tol * N * (it ** 2)) and (max_iter > it)
            l_last = l_curr
        dic = {
                'w': w,
                'means': means,
                'stds': stds,
                'iters': it, 
                'log_like': l_curr - np.log(2 * np.pi) * (d / 2), 
                'сlass': np.argmax(coefs, axis = 1)
              }
        answer.append(dic)
        if l_curr > best['log_like']:
            best = dic
    return {'all': answer, 'best': best}
```

##### Убедимся, что алгоритм удовлетворяет требованиям по скорости работы
(так как проверяем только временные характеристики, то начальное приближение и массив элементов выберем самым простым способом):

```python
N = 10000
d = 100
g = 10
X = np.random.rand(N, d)
w = np.ones(g) / g
means = np.random.rand(g, d)
stds = np.random.rand(g, d, d) / 100 + np.eye(d) / 2
vals = True
for i in range(g):
    stds[i] += stds[i].T
    vals = vals & np.all(np.linalg.eigvals(stds[i]) > 0) # проверим, что положительные 
print('all matrix symmetric and positive: ', vals)
start = [{
            'w' : w,
            'means' : means,
            'stds' : stds
        }]
```

```python
t = - time.time()
ans = EM(X, start, tol = 1e-2)
t = t + time.time()
print('average time of 1 iteration in seconds: ', t / ans['best']['iters'] )
```

Алгоритм незначительно медленнее работает требуемой скорости.  

### 2. Провести тестирование реализованного ЕМ-алгоритма на двумерных модельных данных.

##### 2.1 Восстановление параметров смеси с помощью ЕМ-алгоритма.


```python
# параметры смеси, заметим, что выборка не сбалансированная
N = 1000
d = 2
g = 2
mean0 = np.array([[5, 5], [-1, -1]], dtype = float)
std0 = np.array([[[2, 0.5], [0.5, 2]],
                 [[4, 0], [0, 1]]])
X1 = np.random.multivariate_normal(mean0[0], std0[0], N // 4)
X2 = np.random.multivariate_normal(mean0[1], std0[1], N // 4 * 3)
X  = np.vstack((X1, X2))

x_min = np.min(X, axis = 0)
x_max = np.max(X, axis = 0)
dx_max = x_max - x_min

# начальное приближение
w = np.ones(g) / g
means = np.random.normal(0, 0.3, (g, d)) * dx_max
stds = np.zeros((g, d, d))
for i in range(g):
    stds[i] = np.eye(d) * dx_max / 6
    
start = [{
            'w': w,
            'means': means,
            'stds': stds
        }]

result_med = EM(X, start, max_iter = 5)['best']
result = EM(X, start)['best']
```

Визуально продемонстрируем исходные данные, начальное приближение, результат при 5 итерациях и результат работы алгоритма:

```python
plt.figure(figsize = (14, 10))
for j in range(4):
    splot = plt.subplot(2, 2, j + 1)
    colors = ['r', 'b']
    
    if j == 0:
        Y_ = np.hstack((np.zeros(N // 4), np.ones(N // 4 * 3)))
        stds = std0
        means = mean0
    elif j == 1:
        Y_ = np.hstack((np.zeros(N // 4), np.ones(N // 4 * 3)))
        stds = start[0]['stds']
        means = start[0]['means']
    elif j == 2:
        Y_ = result_med['сlass']
        stds = result_med['stds']
        means = result_med['means']
    elif j == 3:
        Y_ = result['сlass']
        stds = result['stds']
        means = result['means']
    for i in range(2):
        v, w = linalg.eigh(stds[i])
        u = w[0] / linalg.norm(w[0])
        plt.scatter(X[Y_ == i, 0], X[Y_ == i, 1], .8, color=colors[i])

        angle = np.arctan(u[1] / u[0])
        angle = 180 * angle / np.pi  
        ell = Ellipse(means[i], v[0], v[1], 180 + angle, color=colors[i])
        ell.set_clip_box(splot.bbox)
        ell.set_alpha(0.4)
        splot.add_artist(ell)
        if j == 0:
            plt.title('Data')
        elif j == 1:
            plt.title('Initial approximation')
        elif j == 2:
            plt.title('Decision of EM algorithm by 5 iterations')
        elif j == 3:
            plt.title('Decision of EM algorithm')
```

На рисунках результаты восстановленных и исходных параметров смесей незначительно отличаются. Однако параметры отличаются уже во 2 знаке после запятой. Это объясняется сравнительно небольшой выборкой и выбором параметра точности.

### 3. Реализовать оценку модели фона с помощью одномерной гауссианы и и протестировать на последовательности pedestrians.

Для удобства опишем все необходимые функции:

1. Функция загрузки видео(серии изображений): _load()_
2. Функция обучения с применением ЕМ-алгоритма: _training()_
3. Функция оценки принадлежности к фону: _predict3()_  (3 -- номер пункта, данная функция для каждого пункта своя)
4. Функция анализа результата по средствам визуализации ошибок первого и второго рода, и AUC для каждого кадра: _graffics()_

```python
## загрузка видео(набора фотографий)
# можно выбирать сегмент. 
def load(path, make_binary = False, from_p = None, to_p = None, xc = None, yc = None, hsv = False):
    """
    path - path to video
    make_binary - make binary if image is color
    from_p - load picture from from_p position after sorting images
    to_p - load picture to to_p position
    xc, yc - coordinates of pixels with interest
    
    """

    pictures_list = os.listdir(path)
    pictures_list.sort()
    if from_p is None:
        from_p = 0
    if to_p is None:
        to_p = len(pictures_list)

    image = sp.ndimage.imread(path + pictures_list[0])
    shape = image.shape
    if xc is None:
        xc, yc = np.meshgrid(np.arange(shape[0]), np.arange(shape[1]))
    elif yc is None:
        raise Error('forgot yc!')
    xc = xc.ravel()
    yc = yc.ravel()
    size = len(xc)
    color = len(shape) == 3
    out3 = not make_binary and color 
    if out3:
        if hsv:
            video = np.empty((to_p - from_p, size, shape[2]))
        else:
            video = np.empty((to_p - from_p, size, shape[2]), dtype = image.dtype)
    else:
        video = np.empty((to_p - from_p, size, 1), dtype = image.dtype)
    j = 0
    for i in range(from_p, to_p):
        image = sp.ndimage.imread(path + pictures_list[i])
        if color and make_binary:
            image = 0.2126 * image[:, :, 0] + 0.7152 * image[:, :, 1] + 0.0722 * image[:, :, 2]
        if out3:
            if hsv:
                image = image / 255
                image = rgb_to_hsv(image)
            video[j, :, :] = image[xc, yc, :] 
        else:
            video[j, :, 0] = image[xc, yc]
        j += 1
    return video


# обучение
def training(train_video, g = 1):
    """
    g - amount of gaussians
    
    """
    shape = train_video[0].shape
    rows = shape[0]
    size = shape[1]
    means = np.zeros(shape)
    if size == 1:
        stds = np.zeros(shape)
    elif size == 3:
        stds = np.zeros((shape[0], size, size))
    # несколько начальных приближений не будем делать, ибо сложного случая не должно быть
    w = np.ones(g) / g
    means0 = np.ones((g,size)) * 128 + np.random.rand(g, size)
    stds0 = np.empty((g, size, size))
    stds0[:, :, :] = np.eye(size) * 3
    start = [{
                'w' : w,
                'means' : means0,
                'stds' : stds0
            }]
    
    for i in range(rows):
        X = train_video[:, i, :]
        ans = EM(X, start)['best']
        means[i] = ans['means']
        stds[i] = ans['stds']
    return means, stds

# предсказание
def predict3(means, stds, test_video, k = 3):
    """
    k - parameter for decision rule : |x - mean| < (k * std)
    
    """
    shape = stds.shape
    stds_trunc = np.maximum(stds, np.ones(shape, dtype = int) * 3) * k
    maska = np.empty(test_video.shape[:3], dtype = np.bool)
    for i in range(test_video.shape[0]):
        maska[i] = np.abs(test_video[i] - means) > stds_trunc
    return maska

#демонстрация видео
def movie(path, maska, from_p, to_p, xc, yc, pixels, from_frame = 0):
    pictures_list = os.listdir(path)
    image = sp.ndimage.imread(path + pictures_list[0])
    shape = image.shape
    video = load(path, False, from_p, to_p)
    frames = video.shape[0]
    video = np.reshape(video, (frames, shape[0], shape[1], 3), order = 'F')
    maska_all = np.zeros((frames, shape[0], shape[1]), dtype = np.bool)
    for i in range(frames):
        j = i + from_frame
        maska_all[i, xc[pixels[j, :, 0]], yc[pixels[j, :, 0]]] = maska[j, pixels[j]] #.reshape(shape).T
    # mv = vs.make_video(video, maska_all)
    return video, maska_all
    

# вывод
def graffics(predict, y_true, from_frame = 1):
    frames = y_true.shape[0]
    if frames != predict.shape[0]:
        raise IndexError('different sizes of frames')
    FP = np.zeros(frames)
    FN = np.zeros(frames)
    ROC = np.zeros(frames)
    pixels = np.empty(y_true.shape, dtype = np.bool)
    for i in range(frames):
        pixels[i] = (y_true[i] == 0) | (y_true[i] == 255)
        y_true[i, y_true[i] == 255] = 1
        size = pixels.size
        FP[i] = np.sum((y_true[i, pixels[i]] == 0) & (y_true[i, pixels[i]] != predict[i, pixels[i]])) #/ size
        FN[i] = np.sum((y_true[i, pixels[i]] == 1) & (y_true[i, pixels[i]] != predict[i, pixels[i]])) #/ size
        try:
            ROC[i] = roc_auc_score(np.ravel(np.bool8(y_true[i, pixels[i]])), np.ravel(predict[i, pixels[i]]))
        except ValueError:
            ROC[i] = 0
    plt.figure(figsize= (14, 5))
    plt.subplot(1, 2, 1)
    plt.plot(np.arange(frames) + from_frame, FP, '-')
    plt.plot(np.arange(frames) + from_frame, FN, '-')
    plt.title('Errors in algorithm')
    plt.ylabel('share')
    plt.xlabel('number of frame')
    plt.legend(['FP', 'FN'], loc='upper right')
    plt.grid(True)
    
    plt.subplot(1, 2, 2)
    plt.plot(np.arange(frames) + from_frame, ROC, '-')
    plt.ylim((-0.01, 1.2))
    plt.title('AUC-ROC')
    plt.ylabel('accuracy')
    plt.legend(['AUC-ROC'], loc='upper left')
    plt.xlabel('number of frame')
    plt.grid(True)
    return FP, FN, ROC, pixels
```

### 4.Реализация оценки модели фона на основе адаптивной одномерной гауссианы.

Напишем адаптивную функцию, решающую принадлежит ли пиксель "переднему плану": 

```python
def predict4(means, stds, test_video, k = 3, par = 0.05):
    """
    k - parameter for decision rule : |x - mean| < (k * std)
    p - parameter of adapting
    
    """
    shape = stds.shape
    ad_means = means.copy()
    stds_trunc = np.maximum(stds, np.ones(shape, dtype = int) * 3) * k
    maska = np.empty(test_video.shape[:3], dtype = np.bool)
    for i in range(test_video.shape[0]):
        maska[i] = np.abs(test_video[i] - means) > stds_trunc
        ad_p = maska[i] == False
        ad_means[ad_p] = par * test_video[i, ad_p] + (1 - par) * means[ad_p]
        stds_trunc[ad_p] = ((test_video[i] - means) ** 2) * par + (1 - par) * stds_trunc[ad_p]
    return maska
```

Выполним те же действия, что и в пункте 3:

```python
train = load(path_input, True, 0, 300, xc, yc)
means, stds = training(train)
train = []

test = load(path_input, True, 301, 1099, xc, yc)
maska = predict4(means, stds, test) # изменение здесь!
test = []

y_true = load(path_ans, from_p= 301, to_p = 1099, xc = xc, yc = yc)
results = graffics(maska, y_true, from_frame=301)
y_true = []
```

### 5. Реализация оценки модели фона с помощью многомерной гауссианы в цветовом пространстве.

##### 5.1 Цветовое пространство RGB

Напишем функцию для принятия решения о принадлежности к переднему плану в случае трех мерном пространстве:

```python
def predict5(means, stds, test_video, k = 3):
    """
    decide that is fone if N(x | means, stds) > N(means + k * stds | means, stds)
    
    """
    shape = stds.shape
    I = np.eye(stds.shape[1])
    L = np.empty(stds.shape)
    for j in range(stds.shape[0]):
        L[j] = np.linalg.cholesky(stds[j, :, :])
        L[j] = sp.linalg.solve_triangular(L[j].T, I)
    maska = np.empty((test_video.shape[0], test_video.shape[1], 1), dtype = np.bool)
    rule = k**4
    for i in range(test_video.shape[0]):
        dX = test_video[i] - means
        r1 = dX * L[:, 0, :]
        r1 = r1 + dX * L[:, 1, :]
        r1 = r1 + dX * L[:, 2, :]
        maska[i, :, 0] =  np.sum(r1 ** 2, axis = 1) > rule
    return maska
```

```python
train = load(path_input, False, 0, 300, xc, yc)
means, stds = training(train)
train = []

test = load(path_input, False, 301, 1099, xc, yc)
maska = predict5(means, stds, test) # изменение здесь!
test = []

y_true = load(path_ans, from_p= 301, to_p = 1099, xc = xc, yc = yc)
results = graffics(maska, y_true, from_frame=301, rule = 3 ** 4)
y_true = []
```

Причина ухудшения заключается в том, что величины (цвета) очень сильно коррелируют(практически одинаковы), из чего получаем почти вырожденную матрицу, а как следствие слабую разницу с чернобелым изображением.

### 6. Запустить алгоритм разделения смеси 3 трёхмерных гауссиан для вычитания фона в последовательности traffic

```python
def training6(train_video, g = 1):
    """
    g - amount of gaussians
    
    """

    shape = train_video[0].shape
    rows = shape[0]
    size = shape[1]
    means = np.zeros((g, rows, size))
    stds = np.zeros((g, rows, size, size))
    # несколько начальных приближений не будем делать, ибо сложного случая не должно быть
    w = np.ones(g) / g
    means0 = np.ones((g, size))
    dX = (np.random.rand(g, size) - 0.5) * 0.2
    stds0 = np.empty((g, size, size))
    stds0[:, :, :] = np.eye(size) * 0.1
    start = [{
                'w' : w,
                'means' : means0,
                'stds' : stds0
            }]
    for i in range(rows):
        X = train_video[:, i, :] 
        start[0]['means'] = means0 * X.mean(axis = 0) + dX
        ans = EM(X, start)['best']
        means[:, i, :] = ans['means'] 
        stds[:, i, :, :] = ans['stds'] 
    return means, stds
```

### Заключение

В данном задание были рассмотренны способы вычитания фона при помощи ЕМ-алгоритма. В рамках первого пунтка не был реализован ЕМ-алгоритм с диагональной матрицей ковариации. На основании сделанных экспериментов в пунтках __2-6__ можно сделать следующие выводы:

1. ЕМ-алгоритм зависит от начальных приближений, поэтому стоит выбирать несколько начальных приближений
2. Простой алгоритм с полутоновым изображением достаточно неплохо определяет фон(при условии, что он статичен)
3. При добавление адаптивности возможно избавится от выделения медленных изменений в фоне(например движение тени с течением времени). Однако, есть и отрицательные стороны применения такой техники: статические объекты, такие как припаркованный автомобиль, перестают выделяться(в задании не тестировалось).
4. В цветовой палитре лучше проводить классификацию в слабо коррелирующих координатах(например HSV), иначе получаем аналог полутоновой классификации.
5. Классификацию с несколькими гауссианами заметно улучшает качество алгоритма при нестатичном фоне. Отрицательном аспектом здесь является заметное понижение скорости вычисления(кроме увеличение операций для одного ЕМ-шага, ухудшается и скорость сходимости). Несмотря на то, что алгоритм с несколькими гауссианами имеет высокую точность, для рассмотренного примера это не самый лучший вариант. На мой взгляд было бы проще сначала стабилизировать изображение,а потом вычислять фон при помощи одной гауссианы.
