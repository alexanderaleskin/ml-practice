Для удобства вынесем функции, которые описаны ниже и необходимы для выполнения работы(делается, чтобы не запутаться):

```python
def loss_func(theta, visible_size, hidden_size, lambda_, sparsity_param, beta, data):
    
    sizes = np.concatenate((np.array([visible_size]), hidden_size, np.array([visible_size])))
    offset = 0
    n = data.shape[0]
    Z = data
    J = 0
    alfa = lambda_ / 2
    
    for i in range(len(sizes) - 1):
        W = theta[offset : offset + sizes[i] * sizes[i + 1]]
        W = W.reshape(sizes[i], sizes[i + 1])
        b = theta[offset + sizes[i] * sizes[i + 1] : offset + (sizes[i] + 1) * sizes[i + 1]]
        offset += (sizes[i] + 1) * sizes[i + 1]
        A = Z.dot(W) + b
        Z = 1 / (1 + np.exp(-A))
        J += alfa * (np.sum(W ** 2))
        if i < len(sizes) - 2:
            p_real = np.sum(Z, axis = 0) / n
            J += beta * np.sum(p * np.log(p / p_real) + (1 - p) * np.log((1 - p) / (1 - p_real)))
             
    J += np.sum(np.sum((Z - data) ** 2, axis = 1)) / (n * 2)
    
    return J

def generate_patches(images, patch_size=8, step = 4):
    """
    Take a color foto
    step 4 pixels

    return tuple of amount of patches per each image and array of patches
    """

    if type(images) != np.ndarray:
        raise TypeError('images must be of numpy.ndarray')
    if images.ndim != 2:
        raise TypeError('images is 2 dimensional')
    
    sz = np.sqrt(images.shape[1] / 3)
    if np.isclose(int(sz), sz):
        shape = (int(sz), int(sz), 3)
    else:
        raise ValueError('length of iamge should be 3 *row^2')

    patch_per_im = ((shape[0] - patch_size) // step) ** 2
    patches = np.empty((images.shape[0], patch_per_im, 3 * patch_size ** 2))
    pics = images.reshape(images.shape[0], shape[0], shape[0], 3)
    
    x = np.arange(0, shape[0] - patch_size, step)
    xx, yy = np.meshgrid(x, x)
    xx = xx.reshape(-1)
    yy = yy.reshape(-1)
    
    for i in range(patch_per_im):
        temp = pics[:, xx[i]: xx[i] + patch_size, yy[i]: yy[i] + patch_size, :]
        patches[:, i, :] = temp.reshape(pics.shape[0], 3 * patch_size ** 2)                                                                        
    return (patch_per_im, patches.reshape(-1, 3 * patch_size ** 2))
    
```

### Часть I. Вывод и анализ работы алгоритма

В этом пункте выведем __формулу для градиента__ функции потерь, а затем проверим на правильность реализованный алгоритм.


Опишем построение градиента функции при помощи __алгоритма обратного распространения ошибки__.

Пусть обучаем автокодировщик с $K$ слоями: на каждом слое $ n_i$ нейронов, где $i = 0, ... , K$ (так как автокодировщик,  то $n_0 = n_K$). Через $ L $ обозначим количество элементов в выборке, $ x^j$ --- $j$-ый элемент выборки, $y^j$ - значение целевой функции на $j$-ом элементе (в нашем случае $x^j = y ^j$), $W^i $ --- матрица весов между $i - 1$ и $i$ слоями, а $b^i$ ---  вектор смещения для нейронов на $i$ слоем. Так как строим разряженный автокодировщик, то параметр разряженности будем обозначать через $ p $.

Тогда общий вид функции потерь вычисляется по формуле:


$$ \begin{array}  JJ(W, b) = J_1(W, b) + J_2(W, b) + J_3(W, b), где \\ J_1(W, b) = \frac{1}{L} \sum_{j = 1}^{L}  \frac{1}{2} || h_{W, b}(x^j) - y^j || ^2 \text{ ,где }h_{W, b}(x^j) - преобразование нейросети\\ J_2(W, b) =  \frac{\lambda}{2} \sum_{k = 1}^{K} \sum_{i = 1}^{n_{k-1}} \sum_{j = 1}^{n_k} (w_{ij}^k) ^ 2 \text{ ,где } \lambda - \text{ параметр весов,}\\ J_3(W, b) =  \beta \sum_{k = 1}^{K - 1} \sum_{j = 1}^{n_k}\bigg[p \log \big(\frac{p}{\hat{p}_j^k}\big) + (1 - p) \log \big(\frac{1 - p}{1 - \hat{p}_j^k}\big) \bigg] \text{ , где } \hat{p} - \text{среднее значение функции активации нейрона} \end{array} $$



Рассмотрим для начала функционал $ J^*(W, b) = J_1(W, b) + J_3(W, b) $. Воспользуемся алгоритмом обратного распрастранения ошибки. Уточним следующие обозначения и свойства функций:

* В качестве функции активации используем сигмоидную функцию: $ f(x) = \frac{1}{1 + e^{-x}} $ 
* $ f'(x) = \big( \frac{1}{1 + e^{-x}} \big)' = \frac{e^{-x}}{(1 + e^{-x})^2} =  \frac{e^{-x} + 1 - 1}{(1 + e^{-x})^2} = \frac{1}{1 + e^{-x}} \big(1 - \frac{1}{1 + e^{-x}} \big) = f(x) \big(1 - f(x) \big) $
* $ a_0 = X, a_i = f(z_i), \text{где } z_i = a_{i - 1} W^i + b^i, i = 1, ... , K $. Под матрицей Х понимаем матрицу выборки размера $ L \times n_0 $. Заметим, что $ a_K = h_{W, b} $.
* $ \hat{p}_j^k = \frac{1}{L} \sum_{r = 1}^{n_k} f(z^k_{rj})  -$ функция активации нейрона



Вычислим градиент фунции:

1. $\sigma^{n_K} = \big(\frac{\partial J^*}{\partial z^{n_k}_{l, j}} \big)_{l, j}= \big( \frac{\partial}{\partial z^{n_k}_{l, j}} \frac{1}{2 L} || Y - h_{W, b}(X) || ^ 2 \big)_{l, j} = - \frac{1}{L}\big( Y -  h_{W, b}(X)\big) f'(z^{n_k}) \text{ , } $

2. $\sigma^{n_i} = \big(\frac{\partial J^*}{\partial z^{n_i}_{l,j}} \big)_{l, j} = \big(\sum_{r = 1}^{n_{i + 1}}  \frac{\partial J_1}{\partial z^{n_{i+1}}_r} \frac{\partial z^{n_{i+1}}_r}{\partial z^{n_i}_j} +  \beta \frac{\partial}{\partial z^{n_i}_{l,j}} \big[p \log \big(\frac{p}{\hat{p}_j^k}\big) + (1 - p) \log \big(\frac{1 - p}{1 - \hat{p}_j^k}\big) \big] \big)_{l, j} = (W^i)^T \sigma^{n_{i + 1}} f'(Z^i) + \frac{1}{L}\beta \big({- \frac{p}{\hat{p}^i} + \frac{1 - p}{1 - \hat{p}^i}} \big) f'(Z^i) = \bigg[(W^i)^T \sigma^{n_{i + 1}} + \frac{\beta}{L} \big({- \frac{p}{\hat{p}^i} + \frac{1 - p}{1 - \hat{p}^i}} \big) \bigg]f'(Z^i) \text{ , } i = 1, ..., K - 1$

3. $ \big(\frac{\partial J^*}{\partial w^{n_i}_{l,j}} \big)_{l, j} = \big(\sum_{r = 1}^{L}  \frac{\partial J^*}{\partial z^{n_i}_{rj}} \frac{\partial z^{n_i}_{rj}}{\partial w^{n_i}_{l,j}} \big)_{l, j} = (a^i) ^ T \times \sigma^{n_i} -$ "часть" градиента по элементам матрицы $ W^i $

4. $ \big(\frac{\partial J^*}{\partial b^{n_i}_{j}} \big)_{j} = \big(\sum_{r = 1}^{L}  \frac{\partial J^*}{\partial z^{n_i}_{rj}} \frac{\partial z^{n_i}_{rj}}{\partial b^{n_i}_{j}} \big)_{j} =  \big(\sum_{r = 1}^{L}  \sigma^{n_i}_{rj}  \big)_{j} -$ "часть" градиента по элементам вектора смещения $ b^i $

Таким образом, используя алгоритм обратного распространения ошибки, находим градиент $\nabla J^*(W, b) $.

Так как $ J(W, b) = J^*(W, b) + J^2(W, b) \text{ , то } \nabla J(W, b) = \nabla J^*(W, b) + \nabla J_2(W, b) $.

Тогда имеем:

* $ \big(\frac{\partial J}{\partial w^{n_i}_{l,j}} \big)_{l, j} = \big(\frac{\partial J^*}{\partial w^{n_i}_{l,j}}   + \frac{\partial J_2}{\partial w^{n_i}_{l,j}} \big)_{l, j} = (a^i) ^ T \times \sigma^{n_i} + \lambda W^i $

* $ \big(\frac{\partial J}{\partial b^{n_i}_{j}} \big)_{j} = \big(\frac{\partial J^*}{\partial b^{n_i}_{j}} \big)_{j} = \big(\sum_{r = 1}^{L}  \sigma^{n_i}_{rj}  \big)_{j} - \text{ так как вообще говоря } J_2(W, b) = J_2(W) \text{ (независит от b).} $

Для начала реализуем функцию для подсчета функции потерь в точке (W, b), где все переменные вытянуты в 1 вектор theta:


```python
def loss_func(theta, visible_size, hidden_size, lambda_, sparsity_param, beta, data):
    
    sizes = np.concatenate((np.array([visible_size]), hidden_size, np.array([visible_size])))
    offset = 0
    n = data.shape[0]
    Z = data
    J = 0
    alfa = lambda_ / 2
    
    for i in range(len(sizes) - 1):
        W = theta[offset : offset + sizes[i] * sizes[i + 1]]
        W = W.reshape(sizes[i], sizes[i + 1])
        b = theta[offset + sizes[i] * sizes[i + 1] : offset + (sizes[i] + 1) * sizes[i + 1]]
        offset += (sizes[i] + 1) * sizes[i + 1]
        A = Z.dot(W) + b
        Z = 1 / (1 + np.exp(-A))
        J += alfa * (np.sum(W ** 2))
        if i < len(sizes) - 2:
            p_real = np.sum(Z, axis = 0) / n
            J += beta * np.sum(p * np.log(p / p_real) + (1 - p) * np.log((1 - p) / (1 - p_real)))
             
    J += np.sum(np.sum((Z - data) ** 2, axis = 1)) / (n * 2)
    
    return J
    
```

```python
def generate_patches(images, patch_size=8, step = 4):
    """
    Take a color foto
    step 4 pixels

    return tuple of amount of patches per each image and array of patches
    """

    if type(images) != np.ndarray:
        raise TypeError('images must be of numpy.ndarray')
    if images.ndim != 2:
        raise TypeError('images is 2 dimensional')
    
    sz = np.sqrt(images.shape[1] / 3)
    if np.isclose(int(sz), sz):
        shape = (int(sz), int(sz), 3)
    else:
        raise ValueError('length of iamge should be 3 *row^2')

    patch_per_im = ((shape[0] - patch_size) // step) ** 2
    patches = np.empty((images.shape[0], patch_per_im, 3 * patch_size ** 2))
    pics = images.reshape(images.shape[0], shape[0], shape[0], 3)
    
    x = np.arange(0, shape[0] - patch_size, step)
    xx, yy = np.meshgrid(x, x)
    xx = xx.reshape(-1)
    yy = yy.reshape(-1)
    
    for i in range(patch_per_im):
        temp = pics[:, xx[i]: xx[i] + patch_size, yy[i]: yy[i] + patch_size, :]
        patches[:, i, :] = temp.reshape(pics.shape[0], 3 * patch_size ** 2)                                                                        
    return (patch_per_im, patches.reshape(-1, 3 * patch_size ** 2))
    
```
